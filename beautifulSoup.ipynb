```python
''' 
1. 安裝 BeautifulSoup - 用來解析HTML結構
- pip install beautifulsoup4
2. 安裝 requests - 解析網頁的HTML程式碼前，要先把HTML程式碼取回來
- pip install requests
'''

import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get(
    "https://www.eztravel.com.tw")
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify()) 
```

```python
result = soup.find("a","ezTag")
print(result)
```

```python
results = soup.find_all("a","ezTag")
print(results)
```
